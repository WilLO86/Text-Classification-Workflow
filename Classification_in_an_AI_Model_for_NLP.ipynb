For this project, two datasets from Kaggle were selected: players.csv (which contains demographic information about users, including their country of residence) and purchased_games.csv (which records the list of games purchased by each user). The main objective is to integrate both data sources to analyze game consumption by country. Specifically, the analysis aims to:

Consumption Segmentation:
Calculate, for each country, the unique number of players, the average number of games purchased per user, and the total number of games acquired. The analysis is limited to the top 20 countries with the highest consumption.

Game Frequency Analysis:
Explore which game titles are purchased most frequently in each of these countries by extracting and counting each individual game, thereby identifying trends and preferences.

Classification with a Pipeline:
Prepare a classification dataset using the cleaned content of the purchase list as the predictor variable and the country of residence as the target label. Classification models (Logistic Regression and SVC) are trained using TF-IDF vectorization to predict the country based on the list of games. This also enables obtaining prediction probabilities for future predictive analysis.

The results include visualizations that display the countries with the highest consumption (both on average and in total) and the most popular game titles in each of these countries. Moreover, classification pipelines are implemented to predict the class (country) of new texts and to retrieve their probabilities, which can be useful for further predictive tasks.

```python
# ------------------------------
# PART 0: Import Libraries and Load Data
# ------------------------------
import numpy as np                              # For numerical operations
import pandas as pd                             # For DataFrame manipulation
import matplotlib.pyplot as plt                 # For plotting graphs
import seaborn as sns                           # For attractive statistical graphs

# Import functions for train-test splitting and machine learning models
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

# Import tools for text vectorization
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer

# Import modules for text processing (regular expressions and NLTK)
import re
import nltk
nltk.download('stopwords')                       # Download stopwords list
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer         # Stemmer algorithm

# ------------------------------
# PART 1: Load Tables and Merge Data
# ------------------------------
# Load players.csv which contains: playerid, nickname, country
df_players = pd.read_csv('players.csv')

# Load purchased_games.csv which contains: playerid, library (list of games)
df_purchased = pd.read_csv('purchased_games.csv')

# Merge the two DataFrames on the common 'playerid' (inner join to keep matching records)
df_merged = pd.merge(df_purchased, df_players, on='playerid', how='inner')
print("First few rows of the merged dataset:")
print(df_merged.head())

# Display the total number of records and the number of null values in each column
print("\nNumber of records:", len(df_merged))
print("\nNull values in each column:")
print(df_merged.isnull().sum())

# ------------------------------
# PART 2: Clean the Text in the 'library' Column
# ------------------------------
# We perform a less aggressive cleaning: remove punctuation and convert to lowercase,
# but we do not remove stopwords to avoid an empty document.
ps = PorterStemmer()
corpus = []

for i in range(len(df_merged)):
    text = df_merged['library'].iloc[i]
    # Remove any character that is not a letter
    text_clean = re.sub('[^a-zA-Z]', ' ', text)
    # Convert text to lowercase
    text_clean = text_clean.lower()
    # Split text into tokens (words)
    text_tokens = text_clean.split()
    # Apply stemming without removing stopwords to avoid empty documents
    text_tokens = [ps.stem(word) for word in text_tokens]
    # Join tokens back into a single string
    text_clean = ' '.join(text_tokens)
    # If the result is empty, fallback to the original lowercase text
    if text_clean.strip() == "":
        text_clean = text.lower()
    corpus.append(text_clean)

# Add the cleaned text as a new column 'library_clean'
df_merged['library_clean'] = corpus
print("\nFirst 5 cleaned texts from 'library':")
print(df_merged['library_clean'].head())

# ------------------------------
# PART 3: Consumption Analysis by Country (Segmentation)
# ------------------------------
# Create a new column 'num_games' that counts the number of games purchased.
# It is assumed that the 'library' column contains game titles separated by commas.
df_merged['num_games'] = df_merged['library'].apply(lambda x: len(x.split(',')) if pd.notnull(x) else 0)

# Group the data by 'country' to calculate:
# - total_players: unique number of players per country
# - avg_games: average number of games purchased per player
# - total_games: total number of games purchased in the country
segmentation = df_merged.groupby('country').agg(
    total_players=('playerid', 'nunique'),
    avg_games=('num_games', 'mean'),
    total_games=('num_games', 'sum')
).reset_index()

# Select the top 20 countries by total number of games purchased
top20_segmentation = segmentation.sort_values(by='total_games', ascending=False).head(20)
print("\nTop 20 countries by total games purchased:")
print(top20_segmentation)

# Visualization: Average number of games purchased per country (Top 20)
plt.figure(figsize=(12,6))
sns.barplot(data=top20_segmentation, x='country', y='avg_games', palette='viridis')
plt.xlabel('Country')
plt.ylabel('Average Number of Games Purchased')
plt.title('Average Number of Games Purchased by Country (Top 20)')
plt.xticks(rotation=45)
plt.show()

# Visualization: Total number of games purchased per country (Top 20)
plt.figure(figsize=(12,6))
sns.barplot(data=top20_segmentation, x='country', y='total_games', palette='mako')
plt.xlabel('Country')
plt.ylabel('Total Number of Games Purchased')
plt.title('Total Number of Games Purchased by Country (Top 20)')
plt.xticks(rotation=45)
plt.show()

# ------------------------------
# PART 4: Analysis of Game Types (Frequency) by Country
# ------------------------------
# Create a new column 'game_list' by splitting the 'library' column using commas and removing extra spaces
df_exploded = df_merged.copy()
df_exploded['game_list'] = df_exploded['library'].apply(
    lambda x: [game.strip() for game in x.split(',')] if pd.notnull(x) else []
)
# Use the explode method so that each game gets its own row
df_exploded = df_exploded.explode('game_list')
# Rename the exploded column to 'game'
df_exploded.rename(columns={'game_list': 'game'}, inplace=True)
print("\nData after exploding the 'library' column:")
print(df_exploded[['playerid', 'country', 'game']].head(10))

# Filter the exploded DataFrame to include only the top 20 countries
top20_countries = top20_segmentation['country'].tolist()
df_exploded_top20 = df_exploded[df_exploded['country'].isin(top20_countries)]

# Group by 'country' and 'game' to count the frequency of each game in each country
game_freq = df_exploded_top20.groupby(['country', 'game']).size().reset_index(name='frequency')
# Sort in descending order to see the most frequent games globally
game_freq_sorted = game_freq.sort_values(by='frequency', ascending=False)
print("\nTop 10 most frequent games globally (Top 20 countries):")
print(game_freq_sorted.head(10))

# Visualization: Plot the top 5 games for each of the top 20 countries
for country in top20_countries:
    top_games = game_freq[game_freq['country'] == country].sort_values(by='frequency', ascending=False).head(5)
    if top_games.empty:
        continue
    plt.figure(figsize=(8,4))
    sns.barplot(data=top_games, x='frequency', y='game', palette='plasma')
    plt.title(f"Top 5 Games in {country}")
    plt.xlabel('Frequency')
    plt.ylabel('Game')
    plt.show()

# ------------------------------
# PART 5: Classification Pipeline
# ------------------------------
# Prepare a classification dataset using 'library_clean' as text and 'country' as the label.
df_final = df_merged[['country', 'library_clean']].copy()
# Rename the columns to 'label' and 'message'
df_final.columns = ['label', 'message']
print("\nFirst few rows of the final classification dataset:")
print(df_final.head())
print("\nNumber of records in the classification dataset:", len(df_final))
print("\nNull values in the classification dataset:")
print(df_final.isnull().sum())

# ------------------------------
# PART 6: Train-Test Split and Text Vectorization
# ------------------------------
# Assign the variables X (text) and y (label)
X = df_final['message']
y = df_final['label']

# Split the dataset into training (70%) and testing (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
print("\nShape of X_train:", X_train.shape)

# Check the content of X_train to ensure it's not empty
print("Examples from X_train:")
print(X_train.head())

# Use CountVectorizer to convert text into a matrix of token counts (Bag of Words)
count_vect = CountVectorizer(stop_words=None)
X_train_counts = count_vect.fit_transform(X_train)
print("Shape of CountVectorizer output:", X_train_counts.shape)

# Use TfidfTransformer to transform the token counts into TF-IDF features
tfidf_transformer = TfidfTransformer()
X_train_tfidf = tfidf_transformer.fit_transform(X_train_counts)
print("Shape of TF-IDF features:", X_train_tfidf.shape)

# ------------------------------
# PART 7: Create and Train the Classification Model
# ------------------------------
# First approach: Train a Logistic Regression model using the manually obtained TF-IDF features
clf = LogisticRegression(solver='lbfgs')
clf.fit(X_train_tfidf, y_train)

# Second approach: Build a pipeline that integrates TF-IDF vectorization and the classifier
from sklearn.pipeline import Pipeline
text_clf = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words=None)),  # Ensure no stopwords removal here
    ('clf', LogisticRegression(solver='lbfgs')),
])
text_clf.fit(X_train, y_train)

# Make predictions on the test set using the pipeline
predictions = text_clf.predict(X_test)

# Evaluate the classifier: Print the confusion matrix, classification report, and accuracy score
print("\nConfusion Matrix\n", metrics.confusion_matrix(y_test, predictions), "\n")
print("Classification Report\n", metrics.classification_report(y_test, predictions), "\n")
print("Accuracy Score:", metrics.accuracy_score(y_test, predictions))

# ------------------------------
# PART 8: Prediction Functions
# ------------------------------
def predict_text(model, text):
    """
    Predict the class of the given text using the trained model.

    :param model: The trained text classification model (pipeline).
    :param text: A string containing the text to classify.
    :return: The predicted class of the text.
    """
    prediction = model.predict([text])
    return prediction[0]

# Example usage of the predict_text function
your_text = "Your sample text here"
print("\nPredicted class:", predict_text(text_clf, your_text))

def predict_text_with_score(model, text):
    """
    Predict the class of the given text using the trained model and provide the probability scores.

    :param model: The trained text classification model (pipeline).
    :param text: A string containing the text to classify.
    :return: The predicted class and a dictionary with the probability scores per class.
    """
    prediction = model.predict([text])
    prediction_proba = model.predict_proba([text])
    class_labels = model.classes_
    proba_scores = {class_labels[i]: prediction_proba[0][i] for i in range(len(class_labels))}
    return prediction[0], proba_scores

# Example usage of the predict_text_with_score function
predicted_class, scores = predict_text_with_score(text_clf, your_text)
print("Predicted class:", predicted_class)
print("Probability Scores:", scores)

# Save the trained model to a file
import joblib
joblib.dump(text_clf, 'text_clf_model.joblib')

# ------------------------------
# PART 9: Alternative Classification Using SVC
# ------------------------------
from sklearn.svm import SVC
SVC_text_clf = Pipeline([
    ('tfidf', TfidfVectorizer(stop_words=None)),
    ('clf', SVC(gamma='auto')),
])
SVC_text_clf.fit(X_train, y_train)
SVC_predictions = SVC_text_clf.predict(X_test)
print("\nConfusion Matrix (SVC)\n", metrics.confusion_matrix(y_test, SVC_predictions), "\n")
print("Classification Report (SVC)\n", metrics.classification_report(y_test, SVC_predictions), "\n")
print("Accuracy Score (SVC):", metrics.accuracy_score(y_test, SVC_predictions))
```

Final Summary:


The confusion matrix reveals that the classifier is overwhelmingly biased toward predicting one dominant class (United States), while most other classes receive almost no predicted instances. This indicates that the model struggles to distinguish among the various countries based on the cleaned game purchase lists. The classification report reinforces this observation by showing very low precision, recall, and F1-scores for nearly all classes except for the dominant one, resulting in an overall accuracy of approximately 23%.

Insights and Model Performance:

Insights:
The model's heavy bias toward one class suggests that the current TF-IDF representation of the cleaned purchase lists does not capture sufficient discriminative information to effectively differentiate between countries. There may also be significant overlap in the textual patterns across countries, compounded by potential class imbalance in the training data.

Model Performance:
With an overall accuracy of around 23% and almost zero performance metrics (precision, recall, and F1-score) for many classes, the current model is underperforming. This poor performance highlights that the features extracted from the game purchase lists are not adequate for reliable classification.

Potential Improvements:

Enhanced Text Representation:
Exploring more advanced techniques such as word embeddings (e.g., Word2Vec, GloVe) or transformer-based models (e.g., BERT) could capture more semantic information and improve feature discrimination.

Addressing Class Imbalance:
Implementing strategies like oversampling underrepresented classes, undersampling dominant classes, or using class weights in the classifier could help balance the dataset and improve performance for less frequent classes.

Hyperparameter Tuning:
Adjusting parameters in both the vectorization step (e.g., n-gram ranges, minimum document frequency) and the classifier (e.g., regularization strength in Logistic Regression or kernel choice in SVC) might yield better results.
